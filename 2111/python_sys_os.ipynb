## 模块

> 扩展阅读：[python模块](https://www.runoob.com/python3/python3-module.html])

- 模块：是一个包含所有你定义的函数和变量的文件，其后缀名是.py。模块可以被别的程序引入，以使用该模块中的函数等功能。

- 引用python文件：
  - `import module_name` 会依次在搜索路径中查找脚本名称并导入。
  - `from ... import ...`
  - `from ... import *`
  - 搜索路径存储在 `sys.path` 变量中，且当前目录位于 sys.path 的第一项，因此会优先导入本地模块。

- 每个模块都有一个`__name__`属性，当其值是`__main__`时，表明该模块自身在运行，否则是被引入。如果要使得模块被引入时，模块中的某一程序块不执行，可以用`__name__`属性来使该程序块仅在该模块自身运行时执行

```python
if __name__ == '__main__':
   print('程序自身在运行')
else:
   print('我来自另一模块')```

- 内置的函数 `dir()` 可以找到模块内定义的所有名称
```python
import torch
dir(torch)
```

## 包

**包**是一种管理 Python 模块命名空间的形式。比如一个模块的名称是 A.B， 表示一个包 A 中的子模块 B 。不用担心不同库之间的模块重名的情况。


**`__init__.py`文件的作用**
1. 目录只有包含一个叫做 `__init__.py` 的文件才会被认作是一个包。
2. `__init__.py` 文件可以为空，或者包含一些初始化代码、`__all__` 变量等。
   1. 如果包定义文件 `__init__.py` 存在一个叫做 `__all__` 的列表变量，那么在使用 from package import * 的时候就把这个列表中的所有名字作为包内容导入。
   2. 如果包定义文件 `__init__.py` 不存在`__all__` 的列表变量，那么在 import * 时，包里面所有的子模块都会被导入进来。


**包的引用**

1. 使用 `from package import item` 这种形式的时候，对应的 `item` 既可以是包里面的子模块（子包），也可以是包里面定义的其他名称，比如函数，类或者变量。
  
2. import 时，首先把 item 当作一个包定义的名称，如果没找到，再试图按照一个模块去导入。如果还没找到，抛出一个 :exc:ImportError 异常。
  
3. 如果使用形如 import item.subitem.subsubitem 这种导入形式，除了最后一项，都必须是包，而最后一项则可以是模块或者是包，但是不可以是类，函数或者变量的名字。（取决于是否在各层级包含了 `__init__.py` 文件）

4. 兄弟模块的导入：使用绝对路径导入。例如在 torch.data.utils.dataloader 中引入 torch.utils，需要写成`from torch import utils`。

   1. 无论是隐式的还是显式的相对导入都是从当前模块开始的。主模块的名字永远是`__main__`。**一个Python应用程序的主模块，应当总是使用绝对路径引用**。
