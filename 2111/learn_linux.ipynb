## 任务2：在目录下创建文件夹、删除文件夹

### Linux 的目录结构

> 扩展阅读：[Linux 的系统目录结构](https://www.runoob.com/linux/linux-system-contents.html)

- `/etc`： 上边也提到了，这个是系统中的配置文件，如果你更改了该目录下的某个文件可能会导致系统不能启动。

- `/bin`, `/sbin`, `/usr/bin`, `/usr/sbin`: 这是系统预设的执行文件的放置目录，比如 ls 就是在 /bin/ls 目录下的。

  - 值得提出的是 /bin、/usr/bin 是给系统用户使用的指令（除 root 外的通用用户），而/sbin, /usr/sbin 则是给 root 使用的指令。

- `/var`： 这是一个非常重要的目录，系统上跑了很多程序，那么每个程序都会有相应的日志产生，而这些日志就被记录到这个目录下，具体在 /var/log 目录下，另外 mail 的预设放置也是在这里。


### Linux 文件目录管理
> 扩展阅读： [Linuex 文件目录管理](https://www.runoob.com/linux/linux-file-content-manage.html)

#### 处理目录的常见命令
- `ls`（英文全拼：list files）: 列出目录及文件名

- `cd`（英文全拼：change directory）：切换目录

- `pwd`（英文全拼：print work directory）：显示目前的目录

- `mkdir`（英文全拼：make directory）：创建一个新的目录

- `rmdir`（英文全拼：remove directory）：删除一个空的目录

- `cp`（英文全拼：copy file）: 复制文件或目录

- `rm`（英文全拼：remove）: 删除文件或目录

- `mv`（英文全拼：move file）: 移动文件与目录，或修改文件与目录的名称


#### 文件内容查看

- `cat`  由第一行开始显示文件内容

- `tac`  从最后一行开始显示，可以看出 tac 是 cat 的倒着写

- `nl`   显示的时候，输出行号

- `more` 一页一页的显示文件内容

- `less` 与 more 类似，但是比 more 更好的是可以往前翻页

- `head` 只看头几行

- `tail` 只看尾巴几行

- `od` 以二进制的方式读取文件内容


## 任务3：在目录下下载文件、阅读文件

### 下载文件

- `wget`： wget命令用来从指定的URL下载文件。wget非常稳定，它在带宽很窄的情况下和不稳定的网络中有很强的适应性，如果由于网络的原因下载失败，wget会不断的尝试，直到整个文件下载完毕。如果是服务器打断下载过程，它会再次连到服务器上从停止的地方继续下载。这对从那些限定了链接时间的服务器上下载大文件非常有用。

```shell
# 下载单个文件
wget https://www.linuxde.net/testfile.zip

# 下载多个文件：先用filelist.txt存储多个下载链接（换行分隔，再执行以下命令下载多个文件）
wget -i filelist.txt

# 下载并指定保存文件名
wget -O test.zip https://www.linuxde.net/testfile.zip

# 限速下载
wget --limit-rate=300k https://www.linuxde.net/testfile.zip

# 限制总下载文件大小（以下限制大于5M的总下载量）
wget -Q5m -i filelist.txt

# 断点续传
wget -c http://www.linuxde.net/testfile.zip

# 后台下载
wget -b http://www.linuxde.net/testfile.zip
# 后台下载时，查看下载进度
tail -f wget-log

# 使用用户代理下载
wget --user-agent="Mozilla/5.0 (Windows; U; Windows NT 6.1; en-US) AppleWebKit/534.16 (KHTML, like Gecko) Chrome/10.0.648.204 Safari/534.16" http://www.linuxde.net/testfile.zip

# 增加重试次数
wget --tries=40 http://www.linuxde.net/testfile.zip

# 下载信息存入日志文件
wget -o download.log http://www.linuxde.net/testfile.zip

# 过滤指定格式的文件
wget --reject=gif ur

# 下载指定格式的文件
wget -r -A.pdf url


# FTP下载
# 使用wget匿名ftp下载
wget ftp-url

# 使用wget用户名和密码认证的ftp下载
wget --ftp-user=USERNAME --ftp-password=PASSWORD url
```

> 扩展阅读：

1. 使用 `&`

```shell
# 最后的 & 表明在后台执行打包命令
tar czf /data/backup.tgz /data/backup &
# 执行上述命令后， ctrl+d 可以退出ui界面（不能手动叉掉窗口）
```

2. 使用 `nohup`
```shell
nohup tar czf /data/backup.tgz /data/backup
# ctrl+z：挂起到后台
bg # 后台执行，可以通过 jobs 命令查看当前在后台执行的命令
# 此时，ctrl+d 或者关闭窗口，进程仍然在后台运行
```

3. 使用 `nohup` + `&`
```shell
nohup tar czf /data/backup.tgz /data/backup &
# 此时，ctrl+d 或者关闭窗口，进程仍然在后台运行
```

## 任务8：使用grep和awk从文件中筛选字符串

### grep 命令

> 扩展阅读：[linux grep 命令](https://www.runoob.com/linux/linux-comm-grep.html)

`grep`：在文件中提取和匹配符合条件的字符串。如果需要模糊匹配则使用正则表达式，正则表达式是包含匹配。

命令格式： `grep [选项] "搜索内容" 文件名`

- 选项：`-i`忽略大小写；`-n` 输出行号；`-v` 反向查找；`-r`递归查找子目录中的文件；`-l` 只打印匹配的文件名；`-c`只打印匹配的行数；`--color=auto` 搜索出的关键字用颜色显示。
- 正则表达式包含匹配：`?` 匹配0到1次；`*` 匹配0或多次；`[]`匹配中括号中的任意一个字符；`-`表示范围，例如`[a-z]`；`[^]`匹配不是中括号中的一个字符，例如`[^0-9]`；`^`匹配行首；`$`匹配行尾。

### sed 命令

`sed`：以行为单位，对数据进行替换、删除、新增、选取等。

命令格式：`sed [选项] "command" 输入文本`

- 选项：`-n`silent 模式；`-e` 直接在指令列模式上进行 sed 的动作编辑；`-f` 将sed写入档案内；`-i` 直接修改读取的档案内容。
- 命令：`a`新增；`c`替换；`d`删除；`i`插入；`p` 打印；`s` 替换。

## 任务9：在目录下创建zip和tar压缩文件，并进行解压

### tar 命令

- 解包：tar zxvf filename.tar
- 打包：tar czvf filename.tar dirname

### gz 命令

**处理 .gz 格式文件**
- 解压：gunzip filename.gz / gzip -d filename.gz
- 压缩：gzip filename

**处理 .tar.gz 和 .tgz 格式文件**

- 解压：tar zxvf filename.tar.gz
- 压缩：tar zcvf filename.tar.gz dirname
- 压缩多个文件：tar zcvf filename.tar.gz dirname1 dirname2 dirname3.....

### bz2 命令

**处理 .bz2 文件**
- 解压：bzip2 -d filename.bz2 / bunzip2 filename.bz2
- 压缩：bzip2 -z filename

**处理 .tar.bz2 文件**

- 解压：tar jxvf filename.tar.bz2
- 压缩：tar jcvf filename.tar.bz2 dirname

### bz 命令
**处理 .bz 文件**
- 解压：bzip2 -d filename.bz / bunzip2 filename.bz
**处理 .tar.bz 文件**
解压：tar jxvf filename.tar.bz

### z命令
**处理 .z 文件**
- 解压：uncompress filename.z
- 压缩：compress filename
**处理 .tar.z 文件**
- 解压：tar zxvf filename.tar.z
- 压缩：tar zcvf filename.tar.z dirname

### zip 命令
- 解压：unzip filename.zip
- 压缩：zip filename.zip dirname

## 任务10：使用find和locate定位文件

### find 命令

> 扩展阅读：[linux find 命令](https://www.runoob.com/linux/linux-comm-find.html)

`find`：在系统中搜索符合条件的文件名，如果需要模糊查询则使用通配符匹配，通配符是完全匹配。

命令格式：`find [path] [expression]`
- `path`：待查找的目录路径。如果是多个路径，用空格分隔；如果未指定路径，默认为当前目录。
- `expression`：指定查询条件
  - `-name`：按文件名查找，支持通配符
  - `-type`：按文件类型查找，可以是 f（普通文件）、d（目录）、l（符号链接）等
  - `-size`：按文件大小查找，支持使用 + 或 - 表示大于或小于指定大小，单位可以是 c（字节）、w（字数）、b（块数）、k（KB）、M（MB）或 G（GB）
  - `-mtime`：按修改时间查找，支持使用 + 或 - 表示在指定天数前或后，days 是一个整数表示天数。
  - `-user`：按文件所有者查找
  - `-group`：按文件所属组查找


### locate 命令

> 扩展阅读：[linux locate 命令](https://www.runoob.com/linux/linux-comm-locate.html)

`locate`：去保存文档和目录名称的数据库内，查找合乎范本样式条件的文档或目录。
